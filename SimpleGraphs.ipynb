# Imports

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import csv
```

# Opening Data Set

```python
# function to read file from csv 
def load_data(csv_file="DataSet.csv"):
    data = [] 
    with open(csv_file, 'r') as file: 
        for line in csv.DictReader(file): 
            data.append(line) 
    return data

# function to write dict to file
def write_data(data, csv_file="DataSet.csv"):
    csv_columns = list(data[0].keys())
    try:
        with open(csv_file, 'w') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
            writer.writeheader()
            for line in data:
                writer.writerow(line)
    except IOError:
        print("I/O error")

data = load_data()
```

```python
def pie_graph(data, labs, size=(15,15)):
    plt.rcParams["figure.figsize"] = size
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.axis('equal')
    ax.pie(data, labels = labs,autopct='%1.2f%%')
    plt.show()
```

# Age Graph

```python
ages = [0]*100
plt.rcParams["figure.figsize"] = (30,15)
for row in data:
    try:
        age=int(row['Age'])
        ages[age] += 1
    except Exception as e:
        print(row['Age'])
objects = range(100)

y_pos = np.arange(len(objects))
plt.bar(y_pos, ages, align='edge', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Amount of People')
plt.title('Ages')
```

# Sex distribution

```python
sexes = [0,0,0,0]
plt.rcParams["figure.figsize"] = (5,5)
for row in data:
    try:
        sex=int(row['Sex'])
        if(sex == 1):
            sexes[0] += 1
        elif(sex == 2):
            sexes[1] += 1
        elif(sex == -8):
            sexes[2] += 1
        elif(sex == -9):
            sexes[3] += 1
    except Exception as e:
        print(row['Sex'])

print(sexes)
sexes[2] = sexes[2] + sexes[3] 
del sexes[3]
labels = ["Male", "Female", "N/A or Other"]
pie_graph(sexes, labels, (10,10))
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(data, labels = labels,autopct='%1.2f%%')
plt.show()
```

```python
find_max = set()
plt.rcParams["figure.figsize"] = (10,8)
for row in data:
    try:
        find_max.add(int(row['Kids']))
    except Exception as e:
        print(row['PersonID'],row['Kids'])

num_of_kids = [0]* (max(find_max)+1)
for row in data:
    try:
        num_of_kids[int(row['Kids'])] += 1
    except Exception as e:
        print(row['PersonID'], row['Kids'])
        
y_pos = np.arange(len(num_of_kids))

plt.bar(y_pos, num_of_kids, align='edge', alpha=0.5)
plt.xticks(y_pos, find_max)
plt.ylabel('Amount of People')
plt.title('Number of Kids')
```

```python
education = [0,0,0,0,0]
plt.rcParams["figure.figsize"] = (10,10)
for row in data:
    try:
        edu=int(row['Educ'])
        if(edu == 1):
            education[0] += 1
        elif(edu == 2):
            education[1] += 1
        elif(edu == 3):
            education[2] += 1
        elif(edu == 4):
            education[3] += 1
        elif(edu == -8):
            education[3] += 1
    except Exception as e:
        print(row['Educ'])

print(education)
education = education[1:-1]
plt.rcParams['font.size'] = 18.0

langs = [ "High School Grad", "Some College", "Graduates"]
plt.rcParams["figure.figsize"] = (10,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(education, labels = langs,autopct='%1.2f%%')
plt.show()
```

```python
Married = [0,0,0,0]
plt.rcParams["figure.figsize"] = (10,10)
for row in data:
    try:
        edu=int(row['Married'])
        if(edu == 1):
            Married[0] += 1
        elif(edu == 2):
            Married[1] += 1
        elif(edu == 3):
            Married[2] += 1
        elif(edu == -8):
            Married[3] += 1
    except Exception as e:
        print(row['Married'])
print(Married)

langs = ["Less Than High School Grad", "High School Grad", "Some College", "Graduates"]
pie_graph(Married, langs)
```
